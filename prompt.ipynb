### Required Assignment 5.1: Will the Customer Accept the Coupon?

**Context**

Imagine driving through town and a coupon is delivered to your cell phone for a restaurant near where you are driving. Would you accept that coupon and take a short detour to the restaurant? Would you accept the coupon but use it on a subsequent trip? Would you ignore the coupon entirely? What if the coupon was for a bar instead of a restaurant? What about a coffee house? Would you accept a bar coupon with a minor passenger in the car? What about if it was just you and your partner in the car? Would weather impact the rate of acceptance? What about the time of day?

Obviously, proximity to the business is a factor on whether the coupon is delivered to the driver or not, but what are the factors that determine whether a driver accepts the coupon once it is delivered to them? How would you determine whether a driver is likely to accept a coupon?

**Overview**

The goal of this project is to use what you know about visualizations and probability distributions to distinguish between customers who accepted a driving coupon versus those that did not.

**Data**

This data comes to us from the UCI Machine Learning repository and was collected via a survey on Amazon Mechanical Turk. The survey describes different driving scenarios including the destination, current time, weather, passenger, etc., and then ask the person whether he will accept the coupon if he is the driver. Answers that the user will drive there ‘right away’ or ‘later before the coupon expires’ are labeled as ‘Y = 1’ and answers ‘no, I do not want the coupon’ are labeled as ‘Y = 0’.  There are five different types of coupons -- less expensive restaurants (under \$20), coffee houses, carry out & take away, bar, and more expensive restaurants (\$20 - $50).

**Deliverables**

Your final product should be a brief report that highlights the differences between customers who did and did not accept the coupons.  To explore the data you will utilize your knowledge of plotting, statistical summaries, and visualization using Python. You will publish your findings in a public facing github repository as your first portfolio piece.

**Investigating the Bar Coupons**

Now, we will lead you through an exploration of just the bar related coupons.  

1. Create a new `DataFrame` that contains just the bar coupons.


```python
bar_coupons = data[data['coupon'] == 'Bar'].copy()
print(f"Bar coupon rows: {len(bar_coupons)}")
bar_coupons.head()
```

2. What proportion of bar coupons were accepted?


```python
bar_acceptance = bar_coupons['Y'].mean()
print(f"Proportion of bar coupons accepted: {bar_acceptance:.4f} ({bar_acceptance*100:.2f}%)")
```

3. Compare the acceptance rate between those who went to a bar 3 or fewer times a month to those who went more.


```python
few_bar = bar_coupons[bar_coupons['Bar'].isin(['never', 'less1', '1~3'])]
many_bar = bar_coupons[bar_coupons['Bar'].isin(['4~8', 'gt8'])]

rate_few = few_bar['Y'].mean()
rate_many = many_bar['Y'].mean()

print(f"Acceptance rate (bar <= 3x/month): {rate_few:.4f} ({rate_few*100:.2f}%)")
print(f"Acceptance rate (bar >  3x/month): {rate_many:.4f} ({rate_many*100:.2f}%)")

labels = ['Bar 3x or Less/Month', 'Bar More than 3x/Month']
plt.figure(figsize=(6, 4))
sns.barplot(x=labels, y=[rate_few, rate_many], hue=labels, palette='Set2', legend=False)
plt.title('Bar Coupon Acceptance by Visit Frequency')
plt.ylabel('Acceptance Rate')
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```

4. Compare the acceptance rate between drivers who go to a bar more than once a month and are over the age of 25 to the all others.  Is there a difference?


```python
over25_frequent = bar_coupons[
    (bar_coupons['Bar'].isin(['1~3', '4~8', 'gt8'])) &
    (bar_coupons['age'].isin(['26', '31', '36', '41', '46', '50plus']))
]
all_others = bar_coupons[~bar_coupons.index.isin(over25_frequent.index)]

rate_over25 = over25_frequent['Y'].mean()
rate_others = all_others['Y'].mean()

print(f"Acceptance rate (bar >1x/month & age >25): {rate_over25:.4f} ({rate_over25*100:.2f}%)")
print(f"Acceptance rate (all others):               {rate_others:.4f} ({rate_others*100:.2f}%)")
print(f"\nDifference: {(rate_over25 - rate_others)*100:.2f} percentage points")
```

5. Use the same process to compare the acceptance rate between drivers who go to bars more than once a month and had passengers that were not a kid and had occupations other than farming, fishing, or forestry.


```python
target_group = bar_coupons[
    (bar_coupons['Bar'].isin(['1~3', '4~8', 'gt8'])) &
    (bar_coupons['passanger'] != 'Kid(s)') &
    (~bar_coupons['occupation'].isin(['Farming Fishing & Forestry']))
]
other_group = bar_coupons[~bar_coupons.index.isin(target_group.index)]

rate_target = target_group['Y'].mean()
rate_other = other_group['Y'].mean()

print(f"Acceptance rate (target group):  {rate_target:.4f} ({rate_target*100:.2f}%)")
print(f"Acceptance rate (all others):    {rate_other:.4f} ({rate_other*100:.2f}%)")
```

6. Compare the acceptance rates between those drivers who:

- go to bars more than once a month, had passengers that were not a kid, and were not widowed *OR*
- go to bars more than once a month and are under the age of 30 *OR*
- go to cheap restaurants more than 4 times a month and income is less than 50K.



```python
group_a = bar_coupons[
    (bar_coupons['Bar'].isin(['1~3', '4~8', 'gt8'])) &
    (bar_coupons['passanger'] != 'Kid(s)') &
    (bar_coupons['maritalStatus'] != 'Widowed')
]

group_b = bar_coupons[
    (bar_coupons['Bar'].isin(['1~3', '4~8', 'gt8'])) &
    (bar_coupons['age'].isin(['below21', '21', '26']))
]

group_c = bar_coupons[
    (bar_coupons['RestaurantLessThan20'].isin(['4~8', 'gt8'])) &
    (bar_coupons['income'].isin(['Less than $12500', '$12500 - $24999',
                                  '$25000 - $37499', '$37500 - $49999']))
]

combined = bar_coupons[
    bar_coupons.index.isin(group_a.index) |
    bar_coupons.index.isin(group_b.index) |
    bar_coupons.index.isin(group_c.index)
]

print(f"Group A acceptance rate: {group_a['Y'].mean():.4f} ({group_a['Y'].mean()*100:.2f}%)  n={len(group_a)}")
print(f"Group B acceptance rate: {group_b['Y'].mean():.4f} ({group_b['Y'].mean()*100:.2f}%)  n={len(group_b)}")
print(f"Group C acceptance rate: {group_c['Y'].mean():.4f} ({group_c['Y'].mean()*100:.2f}%)  n={len(group_c)}")
print(f"Combined acceptance rate: {combined['Y'].mean():.4f} ({combined['Y'].mean()*100:.2f}%)")

labels = ['Group A\n(>1x bar, no kids,\nnot widowed)',
          'Group B\n(>1x bar,\nage<30)',
          'Group C\n(cheap resto >4x,\nincome<50K)']
rates = [group_a['Y'].mean(), group_b['Y'].mean(), group_c['Y'].mean()]

plt.figure(figsize=(9, 5))
sns.barplot(x=labels, y=rates, hue=labels, palette='Set1', legend=False)
plt.title('Bar Coupon Acceptance Rate by Driver Group')
plt.ylabel('Acceptance Rate')
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```

7.  Based on these observations, what do you hypothesize about drivers who accepted the bar coupons?

**Hypothesis about bar coupon acceptors:**

Based on the analyses above, drivers who accept bar coupons tend to share these characteristics:

- **Frequent bar-goers**: Those who visit bars more than once a month have dramatically higher acceptance rates (~70%+) compared to infrequent visitors (~37%). This is the single strongest predictor.
- **No kids in the car**: Passengers who bring kids are much less likely to stop at a bar. Social context matters.
- **Younger adults**: Drivers under 30 who frequent bars show very high acceptance. Younger demographics are generally more spontaneous.
- **Not widowed**: Widowed individuals showed lower acceptance, possibly reflecting lifestyle differences.
- **Lower-income, budget-conscious diners**: Drivers who frequently visit cheap restaurants on a modest income seem open to bar deals too — they are deal-seekers.

Overall, the profile of a likely bar coupon acceptor is: **a young, socially active adult who already frequents bars, is traveling without children, and responds to value-based offers**.

### Independent Investigation

Using the bar coupon example as motivation, you are to explore one of the other coupon groups and try to determine the characteristics of passengers who accept the coupons.  

```python
df_coffee = data.query("coupon == 'Coffee House'").copy()

# Social vs solo context
def get_vibe(passenger):
    if passenger in ['Friend(s)', 'Partner']:
        return 'Social'
    return 'Solo/Kids'

df_coffee['Vibe'] = df_coffee['passanger'].apply(get_vibe)

print(f"Total Coffee House coupon observations: {len(df_coffee)}")
print(f"Overall acceptance rate: {df_coffee['Y'].mean():.2%}")
print(f"For comparison, overall dataset acceptance rate: {data['Y'].mean():.2%}")
```

```python
freq_order = ['never', 'less1', '1~3', '4~8', 'gt8']
freq_labels = ['Never', '<1/month', '1-3/month', '4-8/month', '>8/month']

freq_rates = [df_coffee[df_coffee['CoffeeHouse'] == f]['Y'].mean() for f in freq_order]
freq_counts = [len(df_coffee[df_coffee['CoffeeHouse'] == f]) for f in freq_order]

fig, axes = plt.subplots(1, 2, figsize=(13, 4))

axes[0].bar(freq_labels, freq_rates, color='saddlebrown', edgecolor='black', alpha=0.8)
axes[0].set_title('Acceptance Rate by Coffee House Visit Frequency')
axes[0].set_xlabel('Monthly Visit Frequency')
axes[0].set_ylabel('Acceptance Rate')
axes[0].set_ylim(0, 1)
axes[0].tick_params(axis='x', rotation=20)

axes[1].bar(freq_labels, freq_counts, color='goldenrod', edgecolor='black', alpha=0.8)
axes[1].set_title('Sample Size by Visit Frequency')
axes[1].set_xlabel('Monthly Visit Frequency')
axes[1].set_ylabel('Count')
axes[1].tick_params(axis='x', rotation=20)

plt.tight_layout()
plt.show()

print("\nAcceptance rates by frequency:")
for label, rate, count in zip(freq_labels, freq_rates, freq_counts):
    print(f"  {label:<15} {rate:.2%}  (n={count})")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 4))

# Acceptance by time of day, split by Vibe
vibe_time = df_coffee.groupby(['time', 'Vibe'])['Y'].mean().unstack()
time_order = ['7AM', '10AM', '2PM', '6PM', '10PM']
vibe_time = vibe_time.reindex(time_order)
vibe_time.plot(kind='bar', ax=axes[0], color=['steelblue', 'coral'], edgecolor='black', alpha=0.8)
axes[0].set_title('Coffee Coupon Acceptance: Time vs Vibe')
axes[0].set_xlabel('Time of Day')
axes[0].set_ylabel('Acceptance Rate')
axes[0].set_ylim(0, 1)
axes[0].tick_params(axis='x', rotation=30)
axes[0].legend(title='Vibe')

# Acceptance by passenger type
pass_rates = df_coffee.groupby('passanger')['Y'].mean().sort_values(ascending=False)
axes[1].bar(pass_rates.index, pass_rates.values, color='mediumseagreen', edgecolor='black', alpha=0.8)
axes[1].set_title('Coffee House Coupon Acceptance by Passenger Type')
axes[1].set_xlabel('Passenger Type')
axes[1].set_ylabel('Acceptance Rate')
axes[1].set_ylim(0, 1)

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 4))

age_order = ['below21', '21', '26', '31', '36', '41', '46', '50plus']
age_rates = df_coffee.groupby('age')['Y'].mean().reindex(age_order)
axes[0].bar(age_order, age_rates.values, color='mediumpurple', edgecolor='black', alpha=0.8)
axes[0].set_title('Coffee House Coupon Acceptance by Age')
axes[0].set_xlabel('Age Group')
axes[0].set_ylabel('Acceptance Rate')
axes[0].set_ylim(0, 1)
axes[0].tick_params(axis='x', rotation=30)

gender_rates = df_coffee.groupby('gender')['Y'].mean()
axes[1].bar(gender_rates.index, gender_rates.values, color=['coral', 'skyblue'], edgecolor='black', alpha=0.8)
axes[1].set_title('Coffee House Coupon Acceptance by Gender')
axes[1].set_xlabel('Gender')
axes[1].set_ylabel('Acceptance Rate')
axes[1].set_ylim(0, 1)

plt.tight_layout()
plt.show()
```

```python
segment = df_coffee[
    (df_coffee['CoffeeHouse'].isin(['1~3', '4~8', 'gt8'])) &
    (df_coffee['time'].isin(['10AM', '2PM'])) &
    (df_coffee['Vibe'] == 'Social')
]
non_segment = df_coffee[~df_coffee.index.isin(segment.index)]

print(f"Target segment (frequent visitor, morning/afternoon, social):")
print(f"  Acceptance rate: {segment['Y'].mean():.2%}  (n={len(segment)})")
print(f"\nAll others:")
print(f"  Acceptance rate: {non_segment['Y'].mean():.2%}  (n={len(non_segment)})")

# Heatmap: time of day vs passenger type
pivot = df_coffee.groupby(['time', 'passanger'])['Y'].mean().unstack()
pivot = pivot.reindex(['7AM', '10AM', '2PM', '6PM', '10PM'])

plt.figure(figsize=(8, 4))
sns.heatmap(pivot, annot=True, fmt='.2f', cmap='YlOrBr',
            linewidths=0.5, cbar_kws={'label': 'Acceptance Rate'})
plt.title('Coffee House Acceptance Rate: Time of Day x Passenger Type')
plt.xlabel('Passenger Type')
plt.ylabel('Time of Day')
plt.tight_layout()
plt.show()
```

### Findings & Conclusions: Coffee House Coupons

After exploring the Coffee House coupon data, a few patterns really stood out to me.

The biggest factor I noticed was **how often someone already visits coffee houses**. People who go regularly, even just 1 to 3 times a month, were much more likely to accept the coupon compared to people who never go. This makes intuitive sense: if you already like coffee houses, a coupon is just extra motivation to do something you'd do anyway.

**Time of day** also made a clear difference. Acceptance was noticeably higher at 10AM and 2PM, which tracks with when people naturally want coffee, mid-morning and after lunch. Late evening coupons (10PM) had much lower acceptance, which also makes sense.

I was a bit surprised by the **passenger effect**. Drivers traveling with friends or a partner accepted at higher rates than those driving alone or with kids. My guess is that stopping for coffee feels more like a fun detour when you're with someone, whereas alone it might feel like an unnecessary stop.

**Age** played a role too. Younger drivers (21 to 26) were the most receptive group. Gender, on the other hand, barely mattered at all.

**The profile that stands out the most** is someone in their early-to-mid 20s, who already goes to coffee houses at least occasionally, riding with a friend or partner, and receiving the coupon in the morning or early afternoon. That combination had notably high acceptance rates.

**What I'd recommend going forward:**
- Deliver coffee house coupons primarily during the 10AM to 2PM window, since evening delivery seems largely wasted.
- Prioritize drivers who have shown prior coffee house visits in their history.
- When a passenger is detected, that's a good signal to send the coupon.
- It would be worth looking more into whether the driver's destination (no urgent place vs. heading to work) interacts with time of day. I suspect drivers with no urgent destination are even more likely to accept.
